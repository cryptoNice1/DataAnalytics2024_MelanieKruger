```python
import pandas as pd, os

os.chdir(r"C:\Users\Melanie\OneDrive - Rensselaer Polytechnic Institute\Desktop\School\24_Spring\Data Analytics\DA_Project\Old Datasets")
heat_gen = pd.read_csv("heat_gen.csv",skiprows=2,encoding="unicode_escape")

EU_countries = "Austria, Belgium, Bulgaria, Croatia, Cyprus, Czechia, Denmark, Estonia, Finland, France, Germany, Greece, Hungary, Ireland, Italy, Latvia, Lithuania, Luxembourg, Malta, Netherlands, Poland, Portugal, Romania, Slovakia, Slovenia, Spain, Sweden".split(", ")
```

```python
heat_gen
```

```python
commercial = heat_gen[heat_gen["Producer Type"] == "Heat (Commercial)"]
commercial = commercial.drop(columns = ["Producer Type"]) 
commercial.rename(columns = {"Country/area":"Country",
                             "Heat generation (TJ)":"Commercial Heat Gen"},inplace=True)
EU_commerical = commercial[commercial["Country"].isin(EU_countries)].sort_values(by = ["Country"])
wut = EU_commerical.pivot_table(index=["Country","Year"], columns='Technology', values='Commercial Heat Gen',aggfunc='sum')
wut.to_csv("heatgen.csv")
```

```python
commercial = heat_gen[heat_gen["Producer Type"] == "Combined Heat and Power (CHP)"]
commercial = commercial.drop(columns = ["Producer Type"]) 
commercial.rename(columns = {"Country/area":"Country",
                             "Heat generation (TJ)":"CHP Heat Gen"},inplace=True)
EU_commerical = commercial[commercial["Country"].isin(EU_countries)].sort_values(by = ["Country"])
wut = EU_commerical.pivot_table(index=["Country","Year"], columns='Technology', values='CHP Heat Gen',aggfunc='sum')
wut.to_csv("CHP_heatgen.csv")
```
