```python
"""time to fill the the missing country"""
all_countries = list(pubinv["Country"].dropna())

for i in range(len(all_countries)-1):
    start_index = pubinv[pubinv["Country"] == all_countries[i]].index[0]
    end_index = pubinv[pubinv["Country"] == all_countries[i+1]].index[0]
    pubinv.loc[start_index:end_index-1, "Country"] = all_countries[i]
    
#fill the last subset
pubinv["Country"] = pubinv["Country"].fillna(all_countries[-1])
```

```python
#time to fill in the technologies
n = len(all_countries)
technologies = list(pubinv[pubinv["Country"] == all_countries[0]]["Technology"].dropna())

for j in range(n):
    for i in range(len(technologies)-1):
        start_index = pubinv[pubinv["Technology"] == technologies[i]].index[j]
        end_index = pubinv[pubinv["Technology"] == technologies[i+1]].index[j]
        pubinv.loc[start_index:end_index-1, "Technology"] = technologies[i]
    
    if j < n:
        #fill the last subset by finding the location of the start of the next country
        last_index = pubinv[pubinv["Technology"] == technologies[0]].index[j+1]
        pubinv.loc[end_index:last_index-1,"Technology"] = technologies[0]
#fill in the last part using fillna
pubinv["Technology"].fillna(technologies[-1],inplace=True)
```
